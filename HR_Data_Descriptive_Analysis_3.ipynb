```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("Employee_Turn_Over_Analysis_3.csv")
df.shape
```

```python
df.isna().sum()
```

```python
# Heat Map

plt.figure(figsize=(12,12))
corr = df.corr()
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, 
            cmap=sns.diverging_palette(220, 10, as_cmap=True), annot=True)
```

```python
# Pair plot comparision of all salary attributes

plt.figure(figsize=(30,50))
g = sns.pairplot(df, vars=["Second Month Gross Salary","Second Month Basic Pay","Second Month OT","Second Month No Pay Days","Second Month Incentive"],hue="Is Terminated", plot_kws = {'alpha': 0.6, 's': 80, 'edgecolor': 'k'})
# Title 
plt.suptitle('Pair Plot of Employee Job Termination', size = 14);
```

```python
# Box plot with salary distribution for Second Month Gross Salary

f, (ax_box, ax_hist) = plt.subplots(2, sharex=True, gridspec_kw={"height_ratios": (.15, .85)})
 
# Add a graph in each part
sns.boxplot(df["Second Month Gross Salary"], ax=ax_box)
sns.distplot(df["Second Month Gross Salary"], ax=ax_hist)
 
# Remove x axis name for the boxplot
ax_box.set(xlabel='')
```

```python
# Box plot for Second Month Gross Salary

g = sns.catplot(x="Is Terminated", y="Second Month Gross Salary",

                data=df, kind="box",

                height=10, aspect=.7);
```

```python
# Box plot for Second Month Basic Pay

g = sns.catplot(x="Is Terminated", y="Second Month Basic Pay",

                data=df, kind="box",

                height=10, aspect=.5);
```

```python
# Box plot for Second Month Incentive

g = sns.catplot(x="Is Terminated", y="Second Month Incentive",

                data=df, kind="box",

                height=10, aspect=.5);
```

```python
# Box plot for Second Month OT

g = sns.catplot(x="Is Terminated", y="Second Month OT",

                data=df, kind="box",

                height=10, aspect=.5);
```

```python
# Scatter plot Second Month Gross Salary over Second Month OT

ax = sns.scatterplot(x="Second Month Gross Salary", y="Second Month OT", hue="Is Terminated", data=df)
```

```python
# Scatter plot Second Month Gross Salary over Second Month Incentive

ax = sns.scatterplot(x="Second Month Gross Salary", y="Second Month Incentive", hue="Is Terminated", data=df)
```

```python
total = df.shape[0]
plt.subplots(figsize=(20,8))
ax = sns.countplot(x='Salary_Category', hue='Is Terminated', data= df, palette = 'colorblind')
ax.set_xlabel('Salary Category')
ax.set_ylabel('Number of Employee')
ax.set_title("Distribution of Employee Termination Over Salary Category")
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2.,
            height + 3,
            '{:1.2f}%'.format(height/total * 100),
            ha="center")
```
